```python
catcols= []
numcols = []
for column in df.columns:
    if df[column].dtype == 'object':
        catcols.append(column)
    elif df[column].dtype in ['int64', 'float64']:
        numcols.append(column)

print(catcols)
print("Number of categorical columns: ", len(catcols))
print()
print(numcols)
print("Number of categorical columns: ", len(numcols))
```

```python
le = {}
for column in ['Gender', 'Attrition_Flag']: 
    le[f'le_{column}'] = LabelEncoder()
    labels = le[f'le_{column}'].fit_transform(df[column])
    df.insert(df.columns.get_loc(column) + 1, f'{column}_l', labels)
print(le) # dictionary: le contains all the fitted encoders
df.head()
```

```python
df.columns
```

```python
education_l=[]
marital_l=[]
income_l=[]
card_l=[]
for i, data in df.iterrows():
    if data['Education_Level'] == 'Unknown':
        education_l.append(0)
    if data['Education_Level'] == 'Uneducated':
        education_l.append(1)
    if data['Education_Level'] == 'High School':
        education_l.append(2)
    if data['Education_Level'] == 'College':
        education_l.append(3)
    if data['Education_Level'] == 'Graduate':
        education_l.append(4)
    if data['Education_Level'] == 'Post-Graduate':
        education_l.append(5)
    if data['Education_Level'] == 'Doctorate':
        education_l.append(6)
        
    if data['Income_Category'] == 'Unknown':
        income_l.append(0)
    if data['Income_Category'] == 'Less than $40K':
        income_l.append(1)
    if data['Income_Category'] == '$40K - $60K':
        income_l.append(2)
    if data['Income_Category'] == '$60K - $80K':
        income_l.append(3)
    if data['Income_Category'] == '$80K - $120K':
        income_l.append(4)
    if data['Income_Category'] == '$120K +':
        income_l.append(5)
        
    if data['Card_Category'] == 'Blue':
        card_l.append(0)
    if data['Card_Category'] == 'Silver':
        card_l.append(1)
    if data['Card_Category'] == 'Gold':
        card_l.append(2)
    if data['Card_Category'] ==  'Platinum':
        card_l.append(3)
 
    if data['Marital_Status'] == 'Unknown':
        marital_l.append(0)
    if data['Marital_Status'] == 'Married':
        marital_l.append(1)
    if data['Marital_Status'] == 'Single':
        marital_l.append(2)
    if data['Marital_Status'] == 'Divorced':
        marital_l.append(3)    
```

```python
df.insert(df.columns.get_loc('Card_Category') + 1, 'card_l', card_l)
df.insert(df.columns.get_loc('Marital_Status') + 1, 'marital_l', marital_l)
df.insert(df.columns.get_loc('Income_Category') + 1, 'income_l', income_l)
df.insert(df.columns.get_loc('Education_Level') + 1, 'edu_l', education_l)

df.columns
```

```python
kmeans_pca = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
```

```python
kmeans_pca.fit(scores_pca)
```

```python
#kmeans clustering with pca result 
# We create a new data frame with the original features and add the PCA scores and assigned clusters.
df_segm_pca_kmeans = pd.concat([df.reset_index(drop = True), pd.DataFrame(scores_pca)], axis = 1)
df_segm_pca_kmeans.columns.values[-2: ] = ['Component 1', 'Component 2']
# The last column we add contains the pca k-means clustering labels.
df_segm_pca_kmeans['Segment K-means PCA'] = kmeans_pca.labels_
```

```python
df_segm_pca_kmeans.head()
```

```python
#calculate mean by segment
df_segm_pca_kmeans_freq = df_segm_pca_kmeans.groupby(['Segment K-means PCA']).mean()
df_segm_pca_kmeans_freq
```

```python
# Calculate the size of each cluster and its proportion to the entire data set.
df_segm_pca_kmeans_freq['N Obs'] = df_segm_pca_kmeans[['Segment K-means PCA','Gender_l']].groupby(['Segment K-means PCA']).count()
df_segm_pca_kmeans_freq['Prop Obs'] = df_segm_pca_kmeans_freq['N Obs'] / df_segm_pca_kmeans_freq['N Obs'].sum()
df_segm_pca_kmeans_freq = df_segm_pca_kmeans_freq.rename({0:'first', 
                                                          1:'second',
                                                          2:'third'})
df_segm_pca_kmeans_freq
```

```python
df_segm_pca_kmeans['Legend'] = df_segm_pca_kmeans['Segment K-means PCA'].map({0:'first', 
                                                          1:'second',
                                                          2:'third'})
```

```python
# Plot data by PCA components. The Y axis is the first component, X axis is the second.
x_axis = df_segm_pca_kmeans['Component 2']
y_axis = df_segm_pca_kmeans['Component 1']
plt.figure(figsize = (10, 8))
sns.scatterplot(x_axis, y_axis, hue = df_segm_pca_kmeans['Legend'], palette = ['g', 'r','c'])
plt.title('Clusters by PCA Components',fontsize=14)
plt.show()
```
